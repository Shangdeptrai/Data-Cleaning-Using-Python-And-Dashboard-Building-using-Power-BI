```python

df = pd.read_excel('/content/messy.xlsx')

# 1. Clean the names of columns
df.columns = df.columns.str.strip().str.lower().str.replace('%', '').str.replace(' ', '_')

# 2. Remove any empty columns
df = df.dropna(axis=1, how='all')

print(df.head())
```

```python
# 3. Change the date column to the same format ‘YYYY-MM-DD’.
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce') # Chuyển đổi lỗi thành NaN

# Chuyển đổi Unix timestamp thành datetime
df['join_date'] = df['join_date'].fillna(pd.to_datetime(df['join_date'], unit='s'))

df['join_date'] = df['join_date'].dt.strftime('%Y-%m-%d')

print(df.head())
```

```python
# 4. Change the name column to title case
df['fll_nam'] = df['fll_nam'].str.title()

print(df.head())
```

```python
# 5. Create the 'email' column
df['email'] = df['fll_nam'].str.split(' ').str[1] + '.' + df['fll_nam'].str.split(' ').str[0] + '.' + df['cust_id'].astype(str) + '@yourcompany.com'

print(df.head())
```

```python
# 6. Change phone number format
df['mobiles'] = '84' + df['mobiles'].astype(str).str.strip().str[-9:]
```

```python
df
```

```python
# 6. Remove duplicated IDs, keeping the earliest join date
df = df.sort_values(by=['cust_id', 'join_date'])
df = df.drop_duplicates(subset='cust_id', keep='first')

print(df.head())
```

```python
# 7. Filter and export to CSV
report_date = date.today().strftime('%Y%m%d')
df_filtered = df[pd.to_datetime(df['join_date']) >= '2019-01-01']
df_filtered.to_csv(f'emp_{report_date}.csv', sep='|', index=False)
```
